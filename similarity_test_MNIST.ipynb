```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from sklearn.datasets import fetch_openml
```

```python
MNIST = fetch_openml('mnist_784', version=1)
```

```python
images = MNIST['data'].to_numpy() / 255.
labels = MNIST['target'].to_numpy().astype(np.int64)
```

```python
@interact(index=(0, len(images)-1))
def show_image(index=0):
    plt.figure(figsize=(4,4))
    plt.imshow(images[index].reshape(28, 28))
    plt.title(f'label: {labels[index]}')
    plt.show()
```

```python
def distance_(x0, x1):
    x = x0 - x1
    distance = np.sqrt(x @ x)
    return distance
```

```python
def angle(x0, x1):
    nominator = x0 @ x1
    denominator = np.sqrt((x0 @ x1) * (x1 @ x1))
    angle = np.arccos(nominator/denominator)
    return angle
```

```python
 distance = []
 for i in range(len(images[:300])):
    for j in range(len(images[:300])):
        distance.append(distance_(images[i], images[i]))
```

```python
len(distance)
```

```python
@interact(first=(0, 299), second=(0, 299), continuous_update=False)
def show_dist_img(first, second):
    plt.figure(figsize=(8,4))
    f = images[first].reshape(28, 28)
    s = images[second].reshape(28, 28)

    ax0 = plt.subplot2grid((2, 2), (0, 0))
    ax1 = plt.subplot2grid((2, 2), (1, 0))
    ax2 = plt.subplot2grid((2, 2), (0, 1), rowspan=2)

    ax0.imshow(f, cmap='gray')
    ax1.imshow(s, cmap='gray')
    ax2.hist(np.array(distance), bins=50)
    d = distance_(f.ravel(), s.ravel())
    ax2.axvline(x=d, ymin=0, ymax=40000, color='C1', linewidth=4)
    ax2.text(0, 16000, "Distance is {:.2f}".format(d), size=12)
    ax2.set(xlabel='distance', ylabel='number of images')
    plt.show()
```

```python
 angles = []
 for i in range(len(images[:300])):
    for j in range(len(images[:300])):
        angles.append(angle(images[i], images[i]))
```

```python
@interact(first=(0, 299), second=(0, 299), continuous_update=False)
def show_dist_img(first, second):
    plt.figure(figsize=(8,4))
    f = images[first].reshape(28, 28)
    s = images[second].reshape(28, 28)

    ax0 = plt.subplot2grid((2, 2), (0, 0))
    ax1 = plt.subplot2grid((2, 2), (1, 0))
    ax2 = plt.subplot2grid((2, 2), (0, 1), rowspan=2)

    ax0.imshow(f, cmap='gray')
    ax1.imshow(s, cmap='gray')
    ax2.hist(np.array(angles), bins=50)
    ang = angle(f.ravel(), s.ravel())
    ax2.axvline(x=ang, ymin=0, ymax=40000, color='C1', linewidth=4)
    ax2.text(0, 16000, "Angle is {:.2f}".format(ang), size=12)
    ax2.set(xlabel='Angle', ylabel='number of images')
    plt.show()
```

```python
def most_similar_dist_image(index):
    distances = np.zeros((300))
    for i in range(300):
        distances[i] = distance_(images[index], images[i])

    most_similar_index = np.where(distances == sorted(distances)[1])[0]
    return most_similar_index
```

```python
@interact(index = (0, 299))
def show_most_similar_distance(index=0):
    plt.figure(figsize=(6, 4))
    f = images[index].reshape(28, 28)
    most_similar_index = most_similar_dist_image(index)[0]
    s = images[most_similar_index].reshape(28, 28)

    ax0 = plt.subplot(121)
    ax1 = plt.subplot(122)

    ax0.imshow(f, cmap='gray')
    ax0.set_title(f'original: {index}')
    ax1.imshow(s, cmap='gray')
    ax1.set_title(f'most_similar: {most_similar_index}')
    plt.show()
```

```python
def most_similar_angle_image(index):
    angles = np.zeros((300))
    for i in range(300):
        angles[i] = angle(images[index], images[i])

    most_similar_index = np.where(angles == sorted(angles)[1])[0]
    return most_similar_index
```

```python
@interact(index = (0, 299))
def show_most_similar_angle(index=0):
    plt.figure(figsize=(6, 4))
    f = images[index].reshape(28, 28)
    most_similar_index = most_similar_angle_image(index)[0]
    s = images[most_similar_index].reshape(28, 28)

    ax0 = plt.subplot(121)
    ax1 = plt.subplot(122)

    ax0.imshow(f, cmap='gray')
    ax0.set_title(f'original: {index}')
    ax1.imshow(s, cmap='gray')
    ax1.set_title(f'most_similar: {most_similar_index}')
    plt.show()
```
